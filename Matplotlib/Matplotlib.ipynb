```python
import matplotlib.pyplot as plt 
import pandas as pd
```

```python
A = [23,25,23,21,22]
B = [20,21,23,24,25]
C = [1,2,3,4,5]

plt.scatter(C,A)

plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Peformance')

plt.show() # show is used to hide <matplotlib.collections.PathCollection at 0x14cdb340610>
```

```python
plt.scatter(C,A, marker = 'D', color = 'Red', label = 'Person A')
plt.scatter(C,B, marker = 'H', color = 'Black', label = 'Person B')
plt.xlabel('Months')
plt.ylabel('Products Sold')
plt.title('Sales Peformance')
plt.grid() # this enable background lines
plt.legend() # to display words mentioned in label parameter
plt.show() # show is used to hide <matplotlib.collections.PathCollection at 0x14cdb340610>
```

```python
plt.subplot(221) # 2 row and 2 columns , 1st graph
plt.scatter(C,A)
plt.ylabel('Y axis')

plt.subplot(222)
plt.plot(C,A) # line plot
plt.xlabel('X axis')
```

```python
plt.subplot(221) # 2 row and 2 columns , 1st graph
plt.scatter(C,A)
plt.ylabel('Y axis')

plt.subplot(222)
plt.plot(C,A) # line plot

plt.subplot(223) # 2 row and 2 columns , 1st graph
plt.scatter(C,B)
plt.ylabel('Y axis')

plt.subplot(224)
plt.plot(C,B) # line plot
plt.xlabel('X axis')

plt.show()
```
