# Matplotlib Basics

```python
x = np.linspace(0, 5, 11)
y = x **2

x
y
```

# Subplots in Matplotlib

```python
# Making sublot of 2/2=4 slots of an canvas(white background board) which has 2 rows & 2 columns ploting slots

# Slot 1 of 4 slots
plt.subplot(2,2,1)  # slot 1 of (2,2)
plt.plot(y,x)

# Slot 2 of 4 slots
plt.subplot(2,2,2)  # slot 2 of (2,2)
plt.plot(x,y)

# Slot 3 of 4 slots  
plt.subplot(2,2,3)  # slot 3 of (2,2)
plt.plot(y**2, x**2)

# Slot 4 of 4 slots 
plt.subplot(2,2,4)  # slot 4 of (2,2)
plt.plot(x**2, y**2)

plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```

# Object Oriented Approach

**When we want to control or customize the size & width of 'canvas'=(behind the white background of plots), we used Object Oriented Way**

```python
# it has full control of canvas for plotting when subplot doesn't have comparetively
fig = plt.figure()  # creating a canvas (white background) for plotting

axis1 = fig.add_axes([0.1, 0.1, 1, 1])  # adding axes with the value of (left margin, bottom margin, width, height) (0.1=10%, 0.8=80%, 1=100%,)
axis1.plot(x,y)

axis2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])  # adding new axes "axis2" which is over lapping the "axis1" (0.3 = 30% left margin, 0.4 = 40% bottom margin)
axis2.plot(x,y)

plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```

**Q. What is the difference between using `subplot()` and the object-oriented approach with `.figure()` in Matplotlib, and when should each be used?**

**Ans:** `subplot()` → A shortcut method that automatically creates both a Figure and Axes in one go, best for quick/simple plots but with limited flexibility in arrangement and styling.

`figure()` (Object Oriented Approach) → Creates only the Figure; you then manually add Axes, giving you full control over layout, positioning, styling, and handling multiple plots in complex, highly customized visualizations.

Matplotlib allows the aspect ratio, DPI (Dots Per Inch) and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.

figsize is a tuple of the width and height of the figure in inches

DPI is the dots-per-inch (pixel per inch). 

DPI = **Dots Per Inch**
In Matplotlib this shows the **resolution** of the plot → how many dots (pixels) will be rendered in one inch.

High DPI → sharper, clearer image.
Low DPI → blurry image.

# Types of Plot

```python
plt.scatter(x,y)  # Scatter Plot. it is a dot-dot plots.
plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```

### Histogram
---------------
* Histogram shows **frequency distribution** of data.
* It works by dividing data into **bins** and counting values per bin.
* Input is **one numeric dataset only** (single variable).
* If you have (x, y) pairs → choose either x **or** y for histogram.
* For relation between x & y → use **scatter/line plot**, not histogram.
* Common use: **visualizing distribution** (shape, spread, skewness).

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)  # Histogram plot
plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```

### Boxplot
---------------
* Boxplot shows **data distribution summary** (min, Q1, median, Q3, max, outliers).
* Input is **one numeric dataset** or **multiple datasets** (list of arrays).
* If you have (x, y) pairs → you still pass **only y values** (distribution).
* Boxplot does **not show relation** between x & y.
* Common use: **compare spread, detect outliers, visualize variability**.

```python
plt.boxplot(x)
plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```
