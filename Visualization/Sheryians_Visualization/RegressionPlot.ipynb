```python
import seaborn as sns 
import matplotlib.pyplot as plt 
import pandas as pd 
import numpy as np 
```

# LM Plot
*regression plot is used in machine learning so we will not fully discuss what, how and where it is used
but we will see how it is created and what are the values we can put inside it*

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips)   # Best fit line 
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex')   # best fit line for 'Male' & 'Female'
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm')   # best fit line for 'Male' & 'Female' with color palette
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o', 'v'])   # best fit line for 'Male' & 'Female' with color palette and marker with 'o' shape and 'v' shape
```

```python
sns.lmplot(x='total_bill', y='tip', data=tips, hue='sex', palette='coolwarm', markers=['o', 'v'], scatter_kws={'s':100})   # best fit line for 'Male' & 'Female' with color palette and marker with 'o' shape and 'v' shape with increased size 100 by using scatter_kws
```
