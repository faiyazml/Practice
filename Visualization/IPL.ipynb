# Sheryians IPL 2022 Visualization 

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

### Matplotlib Basics

```python
x = np.linspace(0, 5, 11)
y = x **3
```

```python
plt.plot(x, y)
plt.title('Visualizing')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

### Subplots in Matplotlib

```python
# Making sublot of 2/2=2 rows 2 columns slots for ploting
plt.title('2/2 Subplot')

# Slot 1 of 4 slots
plt.subplot(2,2,1)  # slot 1 of (2,2)
plt.plot(y,x)
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Slot 2 of 4 slots
plt.subplot(2,2,2)  # slot 2 of (2,2)
plt.plot(x,y)
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Slot 3 of 4 slots  
plt.subplot(2,2,3)  # slot 3 of (2,2)
plt.plot(y**2, x**2)
plt.xlabel('X axis')
plt.ylabel('Y axis')

# Slot 4 of 4 slots 
plt.subplot(2,2,4)  # slot 4 of (2,2)
plt.plot(x**3, y**4)
plt.xlabel('X axis')
plt.ylabel('Y axis')

plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```
