```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
```

## Matplotlib Basics

```python
x = np.linspace(0, 5, 11)
y = x **2
```

```python
plt.plot(x, y)
plt.title('Visualizing')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.show()
```

## Subplots in Matplotlib

```python
# Making sublot of 2/2=4 slots of an canvas(white background board) which has 2 rows & 2 columns ploting slots


# Slot 1 of 4 slots
plt.subplot(2,2,1)  # slot 1 of (2,2)
plt.plot(y,x)

# Slot 2 of 4 slots
plt.subplot(2,2,2)  # slot 2 of (2,2)
plt.plot(x,y)

# Slot 3 of 4 slots  
plt.subplot(2,2,3)  # slot 3 of (2,2)
plt.plot(y**2, x**2)

# Slot 4 of 4 slots 
plt.subplot(2,2,4)  # slot 4 of (2,2)
plt.plot(x**2, y**2)

plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```

## Object Oriented Approach

**When we want to control or customize the size & width of 'canvas'=(behind the white background of plots), we used Object Oriented Way**

```python
# it has full control of canvas for plotting when subplot doesn't have comparetively
fig = plt.figure()  # creating a canvas (white background) for plotting
axis1 = fig.add_axes([0.1, 0.1, 1, 1])  # adding axes with the value of (left margin, bottom margin, width, height) (0.1=10%, 0.8=80%, 1=100%,)
axis1.plot(x,y)
axis2 = fig.add_axes([0.3, 0.4, 0.3, 0.3])  # adding new axes "axis2" which is over lapping the "axis1" (0.3 = 30% left margin, 0.4 = 40% bottom margin)
axis2.plot(x,y)

plt.show()  # to hide the unnecessary text output before or above the Plot Chart
```

**Q. What is the difference between using `subplot()` and the object-oriented approach with `.figure()` in Matplotlib, and when should each be used?**

**Ans:** `subplot()` → A shortcut method that automatically creates both a Figure and Axes in one go, best for quick/simple plots but with limited flexibility in arrangement and styling.

`figure()` (Object Oriented Approach) → Creates only the Figure; you then manually add Axes, giving you full control over layout, positioning, styling, and handling multiple plots in complex, highly customized visualizations.
