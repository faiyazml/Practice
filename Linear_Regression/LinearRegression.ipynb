```python
import numpy as np
import matplotlib.pyplot as plt

# Sample data (Ad Spend in ₹1000, Sale in ₹1000)
x = np.array([1, 2, 3, 4, 5])  # Ad spend
y = np.array([10, 20, 25, 35, 50])  # Sales

# Linear regression using least squares
m, c = np.polyfit(x, y, 1)  # 1 = degree of polynomial (linear)

# Best fit line
y_pred = m * x + c

# Plotting
plt.scatter(x, y, color='blue', label='Actual Data Points')
plt.plot(x, y_pred, color='red', label='Best Fit Line')

# Labels and title
plt.xlabel('Ad Spend (₹1000s)')
plt.ylabel('Sales (₹1000s)')
plt.title('Linear Regression: Best Fit Line')
plt.legend()
plt.grid(True)
plt.show()
```
